## Transverse and Longitudinal Vector Components:

### Direct:
$$
\begin{align}
\hat{\mathbf{z}} \times \left(\mathbf{A} \times \hat{\mathbf{z}}\right) &= \mathbf{A}_t \\
\hat{\mathbf{z}} \cdot \mathbf{A} &= A_z
\end{align}
$$

### Curl:
$$
\begin{align}
\hat{\mathbf{z}} \times \left(\left(\nabla \times \mathbf{A}\right) \times \hat{\mathbf{z}}\right) &= \hat{\mathbf{z}} \times \left(\frac{\partial}{\partial z} \mathbf{A}_t - \nabla_t A_z\right) \\
\hat{\mathbf{z}} \cdot \left(\nabla \times \mathbf{A}\right) &= \hat{\mathbf{z}} \cdot \left(\nabla_t \times \mathbf{A}_t\right) \\
\hat{\mathbf{z}} \times \left(\left(\nabla \times \left(\nabla \times \mathbf{A}\right)\right) \times \hat{\mathbf{z}} \right) &= -\nabla^2 \mathbf{A}_t + \nabla_t\left( \nabla \cdot \mathbf{A} \right) \\
\hat{\mathbf{z}} \cdot \left(\nabla \times \left(\nabla \times \mathbf{A}\right)\right) &= \hat{\mathbf{z}} \cdot \left(\frac{\partial}{\partial z} \left(\nabla_t \cdot \mathbf{A}_t\right) \hat{\mathbf{z}} -\nabla_t^2 A_z \ \hat{\mathbf{z}}\right)
\end{align}
$$

### Gradient:
$$
\begin{align}
\hat{\mathbf{z}} \times \left(\nabla a \times \hat{\mathbf{z}}\right) &= \nabla_t a  \\
\hat{\mathbf{z}} \cdot \nabla a &= \hat{\mathbf{z}} \cdot \left(\frac{\partial}{\partial z} a \ \hat{\mathbf{z}}\right) \\
\hat{\mathbf{z}} \times \left(\nabla \left(\nabla \cdot \mathbf{A}\right) \times \hat{\mathbf{z}}\right) &= \nabla_t \left(\nabla \cdot \mathbf{A}\right)   \\
\hat{\mathbf{z}} \cdot \nabla \left(\nabla \cdot \mathbf{A}\right) &= \hat{\mathbf{z}} \cdot \left(\frac{\partial}{\partial z} \left(\nabla \cdot \mathbf{A}\right) \hat{\mathbf{z}}\right)
\end{align}
$$

### With Matrix Multiplication:
$$
\begin{align}
\epsilon &= \left(
    \begin{array}{ccc}
    \epsilon_{xx} & \epsilon_{xy} & 0 \\
    \epsilon_{yx} & \epsilon_{yy} & 0 \\
    0 & 0 & \epsilon_{zz} \\
    \end{array}\right) \\
\epsilon_t &= \left(
    \begin{array}{ccc}
    \epsilon_{xx} & \epsilon_{xy} & 0 \\
    \epsilon_{yx} & \epsilon_{yy} & 0 \\
    0 & 0 & 0 \\
    \end{array}\right) \\
\end{align}
$$

$$
\begin{align}
\hat{\mathbf{z}} \times \left(\left(\epsilon \cdot \mathbf{A}\right) \times \hat{\mathbf{z}}\right) &= \epsilon_t \cdot \mathbf{A}_t \\
\hat{\mathbf{z}} \cdot \left(\epsilon \cdot \mathbf{A}\right) &= \hat{\mathbf{z}} \cdot \left(\epsilon_{zz} \ A_z \ \hat{\mathbf{z}} \right)
\end{align}
$$

Assuming $\frac{\partial}{\partial z} \epsilon = 0$,
$$
\begin{align}
\hat{\mathbf{z}} \times \left(\epsilon \cdot \left(\mathbf{A}_t \times \hat{\mathbf{z}}\right)\right) &= \mathbf{\tilde{\epsilon}}_t \cdot \mathbf{A}_t \\
\mathbf{\tilde{\epsilon}} &= \left(
    \begin{array}{ccc}
    \epsilon_{yy} & -\epsilon_{yx} & 0 \\
    -\epsilon_{xy} & \epsilon_{xx} & 0 \\
    0 & 0 & \epsilon_{zz} \\
    \end{array}\right) \\
\mathbf{\tilde{\epsilon}}_t &= \left(
    \begin{array}{ccc}
    \epsilon_{yy} & -\epsilon_{yx} & 0 \\
    -\epsilon_{xy} & \epsilon_{xx} & 0 \\
    0 & 0 & 0 \\
    \end{array}\right) = \left(
    \begin{array}{ccc}
    0 & -1 & 0 \\
    1 & 0 & 0 \\
    0 & 0 & 0 \\
    \end{array}\right) \cdot \epsilon \cdot \left(
    \begin{array}{ccc}
    0 & 1 & 0 \\
    -1 & 0 & 0 \\
    0 & 0 & 0 \\
    \end{array}\right) \\
\mathbf{\tilde{\epsilon}}^{-1}_t &= \left(
    \begin{array}{ccc}
    0 & -1 & 0 \\
    1 & 0 & 0 \\
    0 & 0 & 0 \\
    \end{array}\right) \cdot \epsilon^{-1} \cdot \left(
    \begin{array}{ccc}
    0 & 1 & 0 \\
    -1 & 0 & 0 \\
    0 & 0 & 0 \\
    \end{array}\right) \\
\epsilon \cdot \epsilon^{-1} &= \mathbf{1} \\
\mathbf{\tilde{\epsilon}} \cdot \mathbf{\tilde{\epsilon}}^{-1} &= \mathbf{1}
\end{align}
$$

$$
\begin{align}
\hat{\mathbf{z}} \times \left(\left(\nabla \times \left(\epsilon^{-1} \cdot \left(\nabla \times \mathbf{A}\right)\right)\right) \times \hat{\mathbf{z}} \right) &= \mathbf{\tilde{\epsilon}}^{-1}_t \cdot \left( -\frac{\partial^2}{\partial z^2} \mathbf{A}_t + \frac{\partial}{\partial z} \nabla_t A_z \right) + \epsilon^{-1}_{zz} \left( -\nabla^2_t \mathbf{A}_t + \nabla_t \left( \nabla_t \cdot \mathbf{A}_t \right)\right) + \left(\nabla_t \epsilon^{-1}_{zz}\right) \times \left( \nabla_t \times \mathbf{A}_t \right) \\
\hat{\mathbf{z}} \cdot \left(\nabla \times \left(\epsilon^{-1} \cdot \left(\nabla \times \mathbf{A}\right)\right)\right) &= {\mathbf{\tilde{\epsilon}}^{-1}_t}_{ij} \left( \frac{\partial}{\partial z} {\nabla_t}_i {\mathbf{A}_t}_j - {\nabla_t}_i {\nabla_t}_j A_z \right) + \left({\nabla_t}_i {\mathbf{\tilde{\epsilon}}^{-1}_t}_{ij}\right) \left( \frac{\partial}{\partial z} {\mathbf{A}_t}_j - {\nabla_t}_j A_z \right)
\end{align}
$$

## Tranverse and Longitudinal Maxwell's Equations
Assuming a source free medium, longitudinal invariance $\frac{\partial}{\partial z} \epsilon = 0$, and $\epsilon$ generally as defined above, Maxwell's equations are separable into transverse and longitudinal components. In the mixed Fourier domain $\left(\omega, \mathbf{x}\right)$, with $\mathbf{D}$ and $\mathbf{B}$ defined as:
$$
\begin{align}
\mathbf{D} &= \epsilon_0 \left(\epsilon \cdot \mathbf{E}\right) \\
\mathbf{B} &= \mu_0 \mathbf{H}
\end{align}
$$

Maxwell's equations:
$$
\begin{align}
0 &= \nabla \cdot \left(\epsilon \cdot \mathbf{E}\right) \\
0 &= \nabla \cdot \mathbf{H} \\
\mathbf{0} &= \nabla \times \mathbf{E} + i \ \omega \ \mu_0 \ \mathbf{H} \\
\mathbf{0} &= \nabla \times \mathbf{H} - i \ \omega \ \epsilon_0 \left(\epsilon \cdot \mathbf{E}\right)
\end{align}
$$

Wave Equations:
$$
\begin{align}
\mathbf{0} &= \nabla \times \left(\nabla \times \mathbf{E} \right) - \frac{\omega^2}{c^2} \left(\epsilon \cdot \mathbf{E} \right) \\
\mathbf{0} &= \nabla \times \left(\epsilon^{-1} \cdot \left(\nabla \times \mathbf{H}\right)\right) - \frac{\omega^2}{c^2} \mathbf{H}
\end{align}
$$

### Divergence Equations:
$$
\begin{align}
0 &= \nabla_t \cdot \left(\epsilon_t \cdot \mathbf{E}_t\right) + \epsilon_{zz} \ \frac{\partial}{\partial z} E_z \\
0 &= \nabla_t \cdot \mathbf{H}_t + \frac{\partial}{\partial z} H_z
\end{align}
$$

### Transverse Curl Equations:
$$
\begin{align}
\mathbf{0} &= \hat{\mathbf{z}} \times \left(\frac{\partial}{\partial z} \mathbf{E}_t - \nabla_t E_z\right) + i \ \omega \ \mu_0 \ \mathbf{H}_t \\
\mathbf{0} &= \hat{\mathbf{z}} \times \left(\frac{\partial}{\partial z} \mathbf{H}_t - \nabla_t H_z\right) - i \ \omega \ \epsilon_0 \left( \epsilon_t \cdot \mathbf{E}_t \right)
\end{align}
$$

### Longitudinal Curl Equations:
$$
\begin{align}
\mathbf{0} &= \nabla_t \times \mathbf{E}_t + i \ \omega \ \mu_0 \ H_z \hat{\mathbf{z}} \\
\mathbf{0} &= \nabla_t \times \mathbf{H}_t - i \ \omega \ \epsilon_0 \ \epsilon_{zz} \ E_z \ \hat{\mathbf{z}}
\end{align}
$$

### Transverse Wave Equations:
$$
\begin{align}
\mathbf{0} &= -\frac{\partial^2}{\partial z^2} \mathbf{E}_t - \nabla^2_t \mathbf{E}_t + \nabla_t \left(\nabla \cdot \mathbf{E}\right) - \frac{\omega^2}{c^2} \left(\epsilon_t \cdot \mathbf{E}_t\right) \\
&= -\frac{\partial^2}{\partial z^2} \mathbf{E}_t - \nabla^2_t \mathbf{E}_t + \nabla_t \left(\nabla_t \cdot \left(\left(\mathbf{1} - \epsilon_{zz}^{-1} \ \epsilon_t \right) \cdot \mathbf{E}_t \right)\right) + \nabla_t \left(\left( \nabla_t \epsilon_{zz}^{-1} \right) \cdot \left(\epsilon_t \cdot \mathbf{E}_t\right)\right) - \frac{\omega^2}{c^2} \left(\epsilon_t \cdot \mathbf{E}_t\right) \\
\\
\mathbf{0} &= \mathbf{\tilde{\epsilon}}^{-1}_t \cdot \left(-\frac{\partial^2}{\partial z^2} \mathbf{H}_t + \frac{\partial}{\partial z} \nabla_t H_z\right) + \epsilon_{zz}^{-1} \left(-\nabla^2_t \mathbf{H}_t + \nabla_t \left(\nabla_t \cdot \mathbf{H}_t\right)\right) + \left(\nabla_t \epsilon_{zz}^{-1}\right) \times \left(\nabla_t \times \mathbf{H}_t\right) - \frac{\omega^2}{c^2} \mathbf{H}_t \\
 &= -\frac{\partial^2}{\partial z^2} \mathbf{H}_t - \epsilon_{zz}^{-1} \ \mathbf{\tilde{\epsilon}}_t \cdot \left(\nabla_t^2 \mathbf{H}_t\right) - \left(\mathbf{1} - \epsilon_{zz}^{-1} \ \mathbf{\tilde{\epsilon}}_t\right) \cdot \nabla_t \left(\nabla_t \cdot \mathbf{H}_t\right) + \mathbf{\tilde{\epsilon}}_t \cdot \left(\left(\nabla_t \epsilon_{zz}^{-1}\right) \times \left(\nabla_t \times \mathbf{H}_t\right)\right) - \frac{\omega^2}{c^2} \left(\mathbf{\tilde{\epsilon}}_t \cdot \mathbf{H}_t\right)
\end{align}
$$
where the second step is obtained by applying the divergence equations.

## Derivation of the Forward and Backward Propagation Equations
The propagation equations are derived by substituting the modal expansion into the curl equations, with the only additional assumption being that $\epsilon$ is separable into transverse and longitudinal components.

As seen in the following results, the evolution of the amplitude is driven by the following effects:
$$
\frac{\partial}{\partial z} a_n = \begin{cases}
\text{linear dispersion} \\
\text{gain or loss} \\
\text{changing mode profiles} \\
\text{free sources and nonlinearity}
\end{cases}
$$

#### Modal Expansion of 3c
Inserting the expansion of [eq:1](nonlinear_optics.ipynb#eq:lm_expansion) into [eq:4c](nonlinear_optics.ipynb#eq:maxwell) of Maxwell's equations in the [Nonlinear Optics](nonlinear_optics.ipynb) section yields:
$$
\begin{align}
\mathbf{0} &= \sum_m \left(
    \nabla \times \left(a_m \ \hat{\mathbf{e}}_m\right)
    + i \ \omega \ \mu_0 \ \left(a_m \ \hat{\mathbf{h}}_m\right) 
\right) + \left(\text{b.p.m.}\right) \tag{s1} \\
&= \sum_m \left(
    a_m \left(\nabla \times \hat{\mathbf{e}}_m\right)
    + \left(\nabla a_m\right) \times \hat{\mathbf{e}}_m
    + a_m \left(i \ \omega \ \mu_0 \ \hat{\mathbf{h}}_m\right) 
\right) + \left(\text{b.p.m.}\right) \tag{s2} \\
&= \sum_m \left(
    a_m \left(
        \hat{\mathbf{z}} \times \left(
            \frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t
            - \nabla_t {\hat{e}_m}_z
        \right)
        + \nabla_t \times {\hat{\mathbf{e}}_m}_t
    \right)
    + \left(\frac{\partial}{\partial z} a_m\right) \ \hat{\mathbf{z}} \times \hat{\mathbf{e}}_m
    + a_m \left(i \ \omega \ \mu_0 \ \hat{\mathbf{h}}_m\right) 
\right) + \left(\text{b.p.m.}\right) \tag{s3} \\
&= \sum_m \left(
    a_m \left(\hat{\mathbf{z}} \times \frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t\right)
    + \left(\frac{\partial}{\partial z} a_m\right) \hat{\mathbf{z}} \times \hat{\mathbf{e}}_m
    + a_m \left(
        \hat{\mathbf{z}} \times \left( - \nabla_t {\hat{e}_m}_z\right)
        + \nabla_t \times {\hat{\mathbf{e}}_m}_t
        + i \ \omega \ \mu_0 \ \hat{\mathbf{h}}_m
    \right) 
\right) + \left(\text{b.p.m.}\right) \tag{s4} \\
&= \sum_m \left(
    a_m \left(\hat{\mathbf{z}} \times \frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t\right)
    + \left(\frac{\partial}{\partial z} a_m\right) \hat{\mathbf{z}} \times \hat{\mathbf{e}}_m
    + a_m \left(\hat{\mathbf{z}} \times \left(i \ \beta_m \ {\hat{\mathbf{e}}_m}_t \right)\right) 
\right) + \left(\text{b.p.m.}\right) \tag{s5} \\
\end{align}
$$
where the last step was accomplished by substitution of [eq:3c](nonlinear_optics.ipynb#eq:local_modes) and [eq:3e](nonlinear_optics.ipynb#eq:local_modes) in the [Nonlinear Optics](nonlinear_optics.ipynb) section.

Expanding the backward propagating modes, and using the forrward/backward mode relationships yields an equation in terms of the spatial distribution of only the forward propagating modes:
$$
\begin{align}
\mathbf{0} &= \sum_m \left(
    a_m \left(\frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t\right)
    + \left(\frac{\partial}{\partial z} a_m\right) {\hat{\mathbf{e}}_m}_t
    + i \ \beta_m \ a_m \ {\hat{\mathbf{e}}_m}_t
\right) + \left(\text{b.p.m.}\right) \tag{s6} \\
& \begin{split} = \sum_m &\left(
    a_m \left(\frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t\right)
    + \left(\frac{\partial}{\partial z} a_m\right) {\hat{\mathbf{e}}_m}_t
    + i \ \beta_m \ a_m \ {\hat{\mathbf{e}}_m}_t
\right) \\
+ &\left(
    a_{-m} \left(\frac{\partial}{\partial z} {\hat{\mathbf{e}}_{-m}}_t\right)
    + \left(\frac{\partial}{\partial z} a_{-m}\right) {\hat{\mathbf{e}}_{-m}}_t
    + i \ \beta_{-m} \ a_{-m} \ {\hat{\mathbf{e}}_{-m}}_t
\right) \end{split} \tag{s7} \\
& \begin{split} = \sum_m &\left(
    a_m \left(\frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t\right)
    + \left(\frac{\partial}{\partial z} a_m\right) {\hat{\mathbf{e}}_m}_t
    + i \ \beta_m \ a_m \ {\hat{\mathbf{e}}_m}_t
\right) \\
+ &\left(
    a_{-m} \left(\frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t\right)
    + \left(\frac{\partial}{\partial z} a_{-m}\right) {\hat{\mathbf{e}}_m}_t
    - i \ \beta_m \ a_{-m} \ {\hat{\mathbf{e}}_m}_t
\right) \end{split} \tag{s8} \\
&= \sum_m \left(
    \left(a_m + a_{-m}\right) \left(\frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t\right)
    + \left(\frac{\partial}{\partial z} \left(a_m + a_{-m}\right)\right) {\hat{\mathbf{e}}_m}_t
    + i \ \beta_m \ \left(a_m - a_{-m}\right) \ {\hat{\mathbf{e}}_m}_t
\right) \tag{s9} \\
\end{align}
$$

Right multiplying by the cross of $\hat{\mathbf{h}}^*_n$ allows application of the orthogonality relationships:
$$
\begin{align}
\mathbf{0} &= \iint_{A_\infty} \sum_m \left(
    \left(a_m + a_{-m}\right) \left(\frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t\right) \times \hat{\mathbf{h}}^*_n
    + \left(\frac{\partial}{\partial z} \left(a_m + a_{-m}\right)\right) \hat{\mathbf{e}}_m \times \hat{\mathbf{h}}^*_n
    + i \ \beta_m \ \left(a_m - a_{-m}\right) \ {\hat{\mathbf{e}}_m}_t \times \hat{\mathbf{h}}^*_n
\right) \cdot \hat{\mathbf{z}} \ dA \tag{s10}\\
&= \frac{\partial}{\partial z} \left(a_n + a_{-n}\right) + i \ \beta_n \ \left(a_n - a_{-n}\right)
+ \sum_m \iint_{A_\infty} \left(a_m + a_{-m}\right) \left(\left(\frac{\partial}{\partial z} {\hat{\mathbf{e}}_m}_t\right) \times {\hat{\mathbf{h}}^*_n}_t\right)  \cdot \hat{\mathbf{z}} \ dA \tag{s11}\\
\end{align}

    + \left(a_m + a_{-m}\right) \left(
        i \ \beta_m \ \hat{\mathbf{e}}_n^* \times {\hat{\mathbf{h}}_m}_t
        + i \ \omega \ \epsilon_0 \ \hat{\mathbf{e}}_n^* \times \left(\hat{\mathbf{z}} \times \left(
            \left(\epsilon_t - \epsilon_t^{\prime}\right) \cdot {\hat{\mathbf{e}}_m}_t
        \right)\right)
    \right)
\right) \cdot \hat{\mathbf{z}} \ dA \tag{s10} \\
-\iint_{A_\infty} \left({\hat{\mathbf{e}}_n^*}_t \cdot {\mathbf{J}_{f}}_t + i \ \omega \ {\hat{\mathbf{e}}_n^*}_t \cdot {\mathbf{P}_\text{NL}}_t\right) dA &= 
\frac{\partial}{\partial z} \left(a_n - a_{-n}\right)
+ i \ \beta_n \ \left(a_n + a_{-n}\right)
+ \sum_m \iint_{A_\infty} \left(
    \left(a_m - a_{-m}\right) \left({\hat{\mathbf{e}}_n^*}_t \times \left(\frac{\partial}{\partial z} {\hat{\mathbf{h}}_m}_t\right)\right) \cdot \hat{\mathbf{z}}
    + \left(a_m + a_{-m}\right) i \ \omega \ \epsilon_0 \ {\hat{\mathbf{e}}_n^*}_t \cdot \left(
        \left(\epsilon_t - \epsilon^{\prime}_t\right) \cdot {\hat{\mathbf{e}}_m}_t
    \right)
\right) dA \tag{s11} \\
\end{align}
$$

### Forward Mode Propagation Equation
The forward mode propagation equation is found by adding the two curl relations ([eq:4c](nonlinear_optics.ipynb#eq:maxwell) + [eq:4d](nonlinear_optics.ipynb#eq:maxwell)) in the [Nonlinear Optics](nonlinear_optics.ipynb) section. 

$$
\begin{align}
& \begin{split}
-\iint_{A_\infty} \left({\hat{\mathbf{e}}_n^*}_t \cdot {\mathbf{J}_{f}}_t + i \ \omega \ {\hat{\mathbf{e}}_n^*}_t \cdot {\mathbf{P}_\text{NL}}_t\right) \ dA &=
2 \left(
    \frac{\partial}{\partial z} a_n
    + i \ \beta_n \ a_n
\right) \\
& \quad + \sum_m \left(a_m + a_{-m}\right) \iint_{A_\infty} \left(
        \frac{\partial {\hat{\mathbf{e}}_m}_t}{\partial z} \times {\hat{\mathbf{h}}^*_n}_t
    \right) \cdot \hat{\mathbf{z}} \ dA \\
& \quad + \sum_m \left(a_m - a_{-m}\right) \iint_{A_\infty} \left(
        {\hat{\mathbf{e}}_n^*}_t \times \frac{\partial {\hat{\mathbf{h}}_m}_t}{\partial z}
    \right) \cdot \hat{\mathbf{z}} \ dA \\
& \quad + \sum_m \left(a_m + a_{-m}\right) \iint_{A_\infty} \left(
    i \ \omega \ \epsilon_0 \ {\hat{\mathbf{e}}_n^*}_t \cdot
        \left(\epsilon_t - \epsilon^{\prime}_t\right) \cdot {\hat{\mathbf{e}}_m}_t
\right) dA
\end{split} \\ \\
& \begin{split}
\frac{\partial}{\partial z} a_n &=
- i \ \beta_n \ a_n & \quad \text{linear dispersion} \\
& \quad - \sum_{m} \left(a_m + a_{-m}\right) \iint_{A_\infty} \frac{1}{2} \left(
    i \ \omega \ \epsilon_0 \ {\hat{\mathbf{e}}_n^*}_t \cdot
        \left(\epsilon_t - \epsilon^{\prime}_t\right) \cdot {\hat{\mathbf{e}}_m}_t
    \right) dA & \quad \text{gain or loss} \\

    \right) \cdot \hat{\mathbf{z}} \ dA & \quad \text{changing mode profiles (f.p.m.)} \\
& \quad - \sum_{m} a_{-m} \iint_{A_\infty} \frac{1}{2} \left(

\end{split}
\end{align}

## Unit Analysis of the Effective Nonlinear Parameter
The effective nonlinear parameter combines multiple effects attributable to the waveguide mode. Through unit analysis and analogy with standard definitions of the coupled wave equations the effective nonlinearity may be written in terms of effective mode areas, refractive indices, and nonlinear susceptibilities. The reasoning for this is only explicitely drawn out for the second order nonlinearity, but it easily extendible to the third order by the same procedure.

For mode $n$ the second order nonlinearity is written as the product of sums over modes $r$ and $s$, with the effective nonlinearity parameter $(g_{n r s}^{\left(2\right)})$ integrating out all transverse spatial dependence:
$$
\frac{\partial}{\partial z} a_n
= \ldots -i \ \omega \sum_{r s} \iint_{-\infty}^{\infty}
    g_{n r s}^{\left(2\right)}\!\left[\nu_1, \nu_2\right] a_r\!\left[\nu_1\right] a_s\!\left[\nu_2\right]
    \delta\!\left[\nu - \left(\nu_1 + \nu_2\right)\right] d\nu_1 \ d\nu_2
$$
$$
\begin{align}
g_{n r s}^{\left(2\right)}\!\left[\nu_1, \nu_2\right] &= \iint_{A_\infty} \frac{1}{2} \epsilon_0 \ \chi^{\left(2\right)}_{i j k}\!\left[\nu_1, \nu_2\right] {{\hat{\mathbf{e}}_n^*}_t}_i\!\left[\nu_1 + \nu_2\right] {\hat{\mathbf{e}}_r}_j\!\left[\nu_1\right] {\hat{\mathbf{e}}_s}_k\!\left[\nu_2\right] dA
\end{align}
$$

From the orthogonality relationship, the spatial dependence of the electric field scales with the following relationship:
$$
\hat{e} \sim \frac{1}{\sqrt{\epsilon_0 \ c \ n_\text{eff} \ A_\text{eff}}}
$$

Replacing that relatiopnship in the above equations allows grouping the units about each amplitude term $\left(a\right)$:
$$
\begin{align}
\frac{\partial}{\partial z} a 
&\sim \ldots -i \ \omega \iint_{-\infty}^{\infty}
    \left(\frac{1}{2}
        \epsilon_0
        \ \frac{A_\text{eff}}{\left(\epsilon_0 \ c \ n_\text{eff} \ A_\text{eff}\right)^{3/2}} \ \chi^{\left(2\right)}
        \right)
    \ a
    \ a
    \ \delta\!\left[\nu - \left(\nu_1 + \nu_2\right)\right] d\nu_1 \ d\nu_2 \\
\frac{\frac{\partial}{\partial z} a}{\sqrt{\epsilon_0 \ c \ n_\text{eff} \ A_\text{eff}}} 
&\sim \ldots - \frac{1}{2} \frac{i \ \omega}{c \ n_\text{eff}} \iint_{-\infty}^{\infty}
    \chi^{\left(2\right)}
    \frac{a}{\sqrt{\epsilon_0 \ c \ n_\text{eff} \ A_\text{eff}}}
    \frac{a}{\sqrt{\epsilon_0 \ c \ n_\text{eff} \ A_\text{eff}}}
    \ \delta\!\left[\nu - \left(\nu_1 + \nu_2\right)\right] d\nu_1 \ d\nu_2
\end{align}
$$

This form is suggestive of the coupled wave eqution, and is equivalent if the terms about $a$ are interpreted to be the root-intensity of the electric field (see chapter 2 of [Nonlinear Optics (Boyd 2020)](https://doi.org/10.1016/C2015-0-05510-1)). Following this logic, the quantities about each amplitude term should be the corresponding quantities of that amplitude's mode. Replacing the mode identifiers and manipulating the equation back to its origin form yields the natural decomposition of the effective nonlinearity:
$$
\begin{align}
\frac{\frac{\partial}{\partial z} a_n}{\sqrt{\epsilon_0 \ c \ {n_\text{eff}}_n \ {A_\text{eff}}}_n} 
&\sim \ldots - \frac{1}{2} \frac{i \ \omega}{c \ {n_\text{eff}}_n} \iint_{-\infty}^{\infty}
    \chi^{\left(2\right)}_{n r s}
    \frac{a_r}{\sqrt{\epsilon_0 \ c \ {n_\text{eff}}_r \ {A_\text{eff}}_r}}
    \frac{a_s}{\sqrt{\epsilon_0 \ c \ {n_\text{eff}}_s \ {A_\text{eff}}_s}}
    \ \delta\!\left[\nu - \left(\nu_1 + \nu_2\right)\right] d\nu_1 \ d\nu_2 \\
\frac{\partial}{\partial z} a_n 
&\sim \ldots -i \ \omega \iint_{-\infty}^{\infty}
    \left(\frac{1}{2}
        \epsilon_0
        \ \frac{A_\text{eff}}{\left(\epsilon_0 \ c \ n_\text{eff} \ A_\text{eff}\right)^{3/2}} \ \chi^{\left(2\right)}
        \right)
    \ a
    \ a
    \ \delta\!\left[\nu - \left(\nu_1 + \nu_2\right)\right] d\nu_1 \ d\nu_2
\end{align}
$$
$$
g_{n r s}^{\left(2\right)} = \frac{1}{2} \sqrt{\frac{{A_\text{eff}}_n}{{A_\text{eff}}_r \ {A_\text{eff}}_s}} \frac{{\chi^{\left(2\right)}_\text{eff}}_{n r s}}{\sqrt{\epsilon_0 \ c^3 \ {n_\text{eff}}_n \ {n_\text{eff}}_r \ {n_\text{eff}}_s}}
$$
where the frequency dependence of each mode is implicitely assumed.

And likewise for the third order nonlinearity:
$$
g_{n q r s}^{\left(3\right)} = \frac{1}{2} \sqrt{\frac{{A_\text{eff}}_n}{{A_\text{eff}}_q \ {A_\text{eff}}_r \ {A_\text{eff}}_s}} \frac{{\chi^{\left(3\right)}_\text{eff}}_{n q r s}}{\epsilon_0 \ c^2 \ \sqrt{{n_\text{eff}}_n \ {n_\text{eff}}_q \ {n_\text{eff}}_r \ {n_\text{eff}}_s}}
$$

## Relationship Between $\chi^{\left(3\right)}$ and $n_2$
The nonlinear Kerr parameter ($n_2$) describes the change in refractive index due to the instantanous, time averaged optical power. For single frequency, linearly polarized light:
$$
\begin{align}
n &= n_0 + n_2 \left<I\right> \\
&= n_0 + n_2 \left( \epsilon_0 \ c \ n_0 \left<E^2\right>\right)
\end{align}
$$

The third order nonlinearity $\left(\chi^{\left(3\right)}\right)$ can be cast in terms of the Kerr parameter by relating the two through the definition of the relative permmittivity $\left(\epsilon\right)$. Following a procedure similar to [Nonlinear Optics (Boyd 2020)](https://doi.org/10.1016/C2015-0-05510-1), a single frequency electric field is written as the sum of its Fourier amplitude $\left(A\right)$. Its instantanous time average is then given by twice the absolute square of that Fourier amplitude:
$$
E\!\left[t\right] = 2 A \cos\left[\omega \ t - \beta \ z\right] = A e^{-i \left(\omega \ t - \beta \ z\right)} + \text{c.c.} \\
\left<E^2\right> = 2 \bigl|A\bigr|^2
$$

The third order nonlinear polarization may also be expanded in terms of this single frequency field. Most importantly, the expansion includes a term proportional to the electric field:
$$
\begin{align}
P_{NL} = \epsilon_0 \ \chi^{\left(3\right)} \ E\left[t\right]^3 &= \epsilon_0 \ \chi^{\left(3\right)} \left(3 \ \bigl|A\bigr|^2 E\!\left[t\right] + \ldots\right)
\end{align}
$$

The relative permmittivity is defined as the collection of terms in the electric displacement $\left(\mathbf{D}\right)$ directly proportional to the electric field, or equivalently as the square of the refractive index:
$$
\begin{align}
\epsilon &= 1 + \chi^{\left(1\right)} + 3 \ \chi^{\left(3\right)} \bigl|A\bigr|^2 + \ldots \\
\\
\epsilon = n^2 &= \left(n_0 + n_2 \left<I\right>\right)^2 \\
&= n_0^2 + 2 \ n_0 \ n_2 \left<I\right> + \ldots \\
&= n_0^2 + 4 \ \epsilon_0 \ c \ n_0^2 \ n_2 \ \bigl|A\bigr|^2 + \ldots
\end{align}
$$

Finally, equating like terms, those with the same power of $A$, yields the following relationships for the electric susceptibilities:
$$
\begin{align}
\chi^{\left(1\right)} &= n_0^2 - 1 \\
\chi^{\left(3\right)} &= \frac{4}{3} \epsilon_0 \ c \ n_0^2 \ n_2
\end{align}
$$
